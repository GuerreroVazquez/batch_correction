```python
import pandas as pd
```

```python
all_abundances = pd.read_csv("/home/karen/Documents/GitHub/Test_DEG_methods-single_cell/batch_correction/RNAseq_All_abundances_unadjusted.csv")

GSE167186 = pd.read_csv("/home/karen/Documents/phd/Data/Expression_data/GSE129643/GSE129643_abundance.csv")
```

```python
expression_GSE167186 = GSE167186.T.set_axis(GSE167186["Gen"],axis=1)
expression_GSE167186=expression_GSE167186.iloc[1:]
expression_GSE167186
```

```python
expression_GSE167186.columns
```

```python
expression_GSE167186["Experiment"] = ["GSE129643"]*len(expression_GSE167186)
expression_GSE167186
```

```python
metadata_GSE167186 = pd.read_csv("/home/karen/Documents/phd/Data/Expression_data/GSE129643/metadata.csv")
metadata_GSE167186
```

```python
metadata_GSE167186.set_index(metadata_GSE167186["Run"], inplace=True)
```

```python
metadata_GSE167186.drop(columns=["Run"], inplace=True)
```

```python
metadata_GSE167186
```

```python
expression_df = expression_GSE167186.merge(metadata_GSE167186[['Age']], left_index=True, right_index=True)
expression_df["Sample.1"]= expression_df.index
```

```python
expression_df=expression_df.rename_axis("Sample")
```

```python
expression_df
```

```python
all_abundances = all_abundances.set_index("Sample")
```

```python
expression_df.columns
```

```python
(all_abundances.columns==expression_df.columns).all()
```

```python
new_all = pd.concat([all_abundances, expression_df])
```
